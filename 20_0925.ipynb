```python
class BCA_1St_SEMESTER:
     def student(self,a,b=0,c=0,d=0):
         return a+b+c+d
BCA=BCA_1St_SEMESTER()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,0,4,-12))
```

```python
#runtime polymorphism/method overriding
class animal:
     def sound (self):
         print("animal makes a sound")
class dog (animal):
     def sound(self):
         print("bow bow")
class cat(animal):
     def sound(self):
         print("meows meows")
class snake(animal):
     def sound(self):
         print("buuuuuussssss")
obj1=animal()
obj2=dog()
obj3=cat()
obj4=snake()

obj2.sound()
obj1.sound()
obj4.sound()
obj3.sound()
```

```python
from abc import ABC, abstractmethod
class vehicle(ABC):
    def sound(self):
        pass
class car(vehicle):
    def sound(self):
        print("i'm driving a car")
obj = car()
obj.sound()
```

```python
from abc import ABC, abstractmethod
class game(ABC):


    def freefire(self):
        print("just playing ")


    def BGMI(self):
        pass

    def LUDO_KING(self):
        print("i'm win the game")

class TEMPLE_RUN (game):

     def BGMI(self):
        print("royal pass completed")
     def chess(self):
        print("i'm a district player")

obj = TEMPLE_RUN()
obj.BGMI()
obj.LUDO_KING()
obj.chess()
obj.freefire()
```
